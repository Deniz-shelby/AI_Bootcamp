# Computer Vision Morphological Transformations

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

## Dilation

```python
# Create a simple 3 by 3 kernel of ones with numpy
kernel = np.ones((3, 3), np.uint8)

kernel
```

```python
!ls img
```

```python
# Load an image, threshold it (binary_inverse) and apply dilation with openCV, try different number of iterations
text = cv2.imread('./img/t1.png')
```

```python
_, thresh = cv2.threshold(text, 128, 255, cv2.THRESH_BINARY_INV)


for i in range(5):
    dilated = cv2.dilate(thresh, kernel, iterations=i+1)
    plt.figure(figsize = (10,10))
    plt.imshow(dilated);
    plt.title(f'Iterations: {i+1}')
```
