```python
acc_mean = df_w_20['android.sensor.accelerometer#mean']
    
```

```python
step_counter = 0
for i in range(len(acc_mean)-1):
    if (i > 0)&(i < len(acc_mean)):
        y = acc_mean[i]
        y_before = acc_mean[i-1]
        y_after = acc_mean[i+1]
        if (y > y_before) & (y> y_after) & (15 < (y_before+y_after)):
            step_counter +=1
            print(step_counter)
```
